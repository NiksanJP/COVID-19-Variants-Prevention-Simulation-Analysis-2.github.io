```python
import pandas as pd
import plotly.io as pio
import plotly.express as px
import plotly
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import plotly.io as pio
pio.renderers.default = 'iframe_connected'
```

```python
df = pd.read_csv('Data/owid-covid-data.csv')
df.head()
```

```python
df.columns
```

```python
cases_vaccinations = df[['date', 'location', 'new_cases', 'total_vaccinations']]
cases_vaccinations = cases_vaccinations[cases_vaccinations['date'] > '2020-11-1']
```

```python
USA = cases_vaccinations[cases_vaccinations['location'] == 'United States']
Israel = cases_vaccinations[cases_vaccinations['location'] == 'Israel']
```

```python
USA.head()
```

```python
compare_fig = make_subplots(rows=1, cols=2, subplot_titles=("United States", "Israel"), specs=[[{"secondary_y": True},{"secondary_y": True}]])

compare_fig.add_trace(
    go.Scatter(x=USA['date'].tolist(), y=USA['new_cases'].tolist(), name='new Cases'),
    secondary_y=True, row=1, col=1
)

compare_fig.add_trace(
    go.Bar(x=USA['date'].tolist(), y=USA['total_vaccinations'].tolist(), name='Vaccinations'),
    secondary_y=False, row=1, col=1
)

compare_fig.add_trace(
    go.Scatter(x=Israel['date'].tolist(), y=Israel['new_cases'].tolist(), name='new Cases'),
    secondary_y=True, row=1, col=2
)

compare_fig.add_trace(
    go.Bar(x=Israel['date'].tolist(), y=Israel['total_vaccinations'].tolist(), name='Vaccinations'),
    secondary_y=False, row=1, col=2
)

compare_fig.update_layout(height=600, width=2000, title_text="Relationship of COVID-19 cases and Vaccinations")
compare_fig.update_yaxes(showgrid=False)
compare_fig.show()
```
